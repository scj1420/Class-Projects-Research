The goal of this project is to train a computer to play tic tac toe. 

The first step was to develop the game so that two human players could play. The state of the game board is stored as a list; this list is then fed into a variety of functions that display the game board, check for winners, allow the placement of markers, etc. Currently, the placement of markers is specified by a pair of numbers separated by a comma. For example, to place a marker in the first row and second column, type 1,2. In the future, it would be preferable to place markers by simply clicking on the correct location. 

Currently, the game does allow a human to play against a computer; however, the computer simply makes random moves. A function that checks to see if a player is about to win has been written; however, this is not currently implemented by the computer. The first step to training the computer to properly play would be to assign a score to every game-board layout. Algorithms could then be written to get the computer to make the move that results in the optimal score.

Let's now write a function that will assign a score to a given board-game layout. If a row/column/diagonal is empty, it will be assigned a score of 0. Likewise, if a row/column/diagonal has at least one of each character (s.t. it's not possible to win in that row/column/diagonal), it will be assigned a zero. If a row/column/diagonal has only one character (e.g. only an 'X') and no other characters, let's assign it a score of +1 if it's a friendly character and -1 if it's an opponent's character. Similarly, if a row/column/diagonal has two of the same character and no other characters, let's assign it a score of +10 if they are friendly characters and a -100 if they are opponent's characters. A row/column/diagonal with three of the same characters will be assigned a score of +/- 100. The overall score for the game_board layout will be the sum of all the rows, columns, and diagonals.

Note that there may be better scoring functions. If we have extra time, we'll come back and try to come up with them.

```python
def game_board_score_v1(game_board_list, friendly_character, opponent_character):
    characters = [friendly_character, opponent_character]
    score = 0
    for i in range(2):
        single_list = [[characters[i], ' ', ' '], [' ', characters[i], ' '], [' ', ' ', characters[i]]]
        double_list = [[characters[i], characters[i], ' '], [characters[i], ' ', characters[i]], [' ', characters[i], characters[i]]]
        triple_list = [[characters[i], characters[i], characters[i]]]
        big_list = [single_list, double_list, triple_list]
        score_list = [[1, 10, 100], [-1, -10, -100]]
        for n in range(3):
            for j in range(3):
                #checks rows
                if [game_board_list[j][k] for k in range(len(game_board_list))] in big_list[n]: # checks rows
                    score += score_list[i][n]
                #checks columns
                if [game_board_list[k][j] for k in range(len(game_board_list))] in big_list[n]: # checks columns
                    score += score_list[i][n]
            # checks \ diagonal
            if [game_board_list[k][k] for k in range(len(game_board_list))] in big_list[n]: 
                score += score_list[i][n]
            # checks / diagonal
            if [game_board_list[k][l] for k in range(len(game_board_list)) \
                for l in range(len(game_board_list) - 1, -1, -1) if k + l == len(game_board_list) - 1] in big_list[n]:
                score += score_list[i][n]
    return score

game = [['X', ' ', ' '],
        [' ', 'O', ' '],
        [' ', 'O', 'X']]   
game_board_score_v1(game, 'O', 'X')
```

We now need to use the score function to get the computer to compute the best move and then to make that move. We'll write a function that takes a given game-board layout and outputs the best move to make, according to our scoring function.

```python
def best_move_v1(game_board_list, friendly_character, opponent_character):
    import copy
    def place_marker(game_board_list, location, character):
        row = int(location[0])
        column = int(location[2])
        game_board_list[row - 1][column - 1] = character
        
        
    def game_board_score_v1(game_board_list, friendly_character, opponent_character):
        characters = [friendly_character, opponent_character]
        score = 0
        for i in range(2):
            single_list = [[characters[i], ' ', ' '], [' ', characters[i], ' '], [' ', ' ', characters[i]]]
            double_list = [[characters[i], characters[i], ' '], [characters[i], ' ', characters[i]], [' ', characters[i], characters[i]]]
            triple_list = [[characters[i], characters[i], characters[i]]]
            big_list = [single_list, double_list, triple_list]
            score_list = [[1, 10, 100], [-1, -10, -100]]
            for n in range(3):
                for j in range(3):
                    if [game_board_list[j][k] for k in range(len(game_board_list))] in big_list[n]: # checks rows
                        score += score_list[i][n]
                    if [game_board_list[k][j] for k in range(len(game_board_list))] in big_list[n]: # checks columns
                        score += score_list[i][n]
                if [game_board_list[k][k] for k in range(len(game_board_list))] in big_list[n]: 
                    score += score_list[i][n]
                if [game_board_list[k][l] for k in range(len(game_board_list)) \
                    for l in range(len(game_board_list) - 1, -1, -1) if k + l == len(game_board_list) - 1] in big_list[n]:
                    score += score_list[i][n]
        return score
        
    
    available_moves = []
    for i in range(len(game_board_list)):
        for j in range(len(game_board_list[i])):
            if game_board_list[i][j] == ' ':
                available_moves.append(str(i + 1) + "," + str(j + 1))  
    scores_list = []
    for k in range(len(available_moves)):
        new_game_board_list = copy.deepcopy(game_board_list)
        place_marker(new_game_board_list, available_moves[k], friendly_character)
        scores_list.append(game_board_score_v1(new_game_board_list, friendly_character, opponent_character))
    score_dictionary = {}
    for score, move in zip(scores_list, available_moves):
        score_dictionary[score] = move
    return score_dictionary[max(list(score_dictionary.keys()))]
    

best_move_v1([['X', ' ', ' '],
              [' ', 'X', 'O'],
              ['O', ' ', ' ']], "X", "O")
```

```python
def best_move_v2(game_board_list, friendly_character, opponent_character):
        import copy
        import random
        def place_marker(game_board_list, location, character):
            row = int(location[0])
            column = int(location[2])
            game_board_list[row - 1][column - 1] = character


        def game_board_score_v1(game_board_list, friendly_character, opponent_character):
            characters = [friendly_character, opponent_character]
            score = 0
            for i in range(2):
                single_list = [[characters[i], ' ', ' '], [' ', characters[i], ' '], [' ', ' ', characters[i]]]
                double_list = [[characters[i], characters[i], ' '], [characters[i], ' ', characters[i]], [' ', characters[i], characters[i]]]
                triple_list = [[characters[i], characters[i], characters[i]]]
                big_list = [single_list, double_list, triple_list]
                score_list = [[1, 10, 100], [-1, -10, -100]]
                for n in range(3):
                    for j in range(3):
                        if [game_board_list[j][k] for k in range(len(game_board_list))] in big_list[n]: # checks rows
                            score += score_list[i][n]
                        if [game_board_list[k][j] for k in range(len(game_board_list))] in big_list[n]: # checks columns
                            score += score_list[i][n]
                    if [game_board_list[k][k] for k in range(len(game_board_list))] in big_list[n]: 
                        score += score_list[i][n]
                    if [game_board_list[k][l] for k in range(len(game_board_list)) \
                        for l in range(len(game_board_list) - 1, -1, -1) if k + l == len(game_board_list) - 1] in big_list[n]:
                        score += score_list[i][n]
            return score


        # create list of available moves
        available_moves = []
        for i in range(len(game_board_list)):
            for j in range(len(game_board_list[i])):
                if game_board_list[i][j] == ' ':
                    available_moves.append(str(i + 1) + "," + str(j + 1))           
        # check to see if a given move allows the opponent a score greater than 15 (i. e. a high likelihood of winning)
        # if so, remove that move from the available moves list
        if int(len(available_moves)) > 1:
            for move in available_moves:
                # print(move, ':')
                new_game_board_list = copy.deepcopy(game_board_list)
                place_marker(new_game_board_list, move, friendly_character)
                opponent_available_moves = []
                for i in range(len(new_game_board_list)):
                    for j in range(len(new_game_board_list[i])):
                        if new_game_board_list[i][j] == ' ':
                            opponent_available_moves.append(str(i + 1) + "," + str(j + 1))
    #           print("Opponent's available moves: ", opponent_available_moves)
                opponent_scores_list = []
                for k in range(len(opponent_available_moves)):
                    final_game_board_list = copy.deepcopy(new_game_board_list)
                    place_marker(final_game_board_list, opponent_available_moves[k], opponent_character)
                    opponent_scores_list.append(game_board_score_v1(final_game_board_list, opponent_character, friendly_character))
                if max(opponent_scores_list) > 15:
                    available_moves.remove(move)

        scores_list = []
        for k in range(len(available_moves)):
            new_game_board_list = copy.deepcopy(game_board_list)
            place_marker(new_game_board_list, available_moves[k], friendly_character)
            scores_list.append(game_board_score_v1(new_game_board_list, friendly_character, opponent_character))
#         print("Available moves: ", available_moves)
#         print("Corresponding scores (looking two moves ahead): ", scores_list) 
        best_moves_indices = [i for i, j in enumerate(scores_list) if j == max(scores_list)]
        best_moves = [available_moves[i] for i in best_moves_indices]
#         print("Best move(s): ", best_moves)
        return random.choice(best_moves)
```
