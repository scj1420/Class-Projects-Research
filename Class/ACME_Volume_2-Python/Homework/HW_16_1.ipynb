## Problem 16.4

```python
def cake_eating(w, beta, N, T, u):
    k_list = []
    v0 = []
    for i in range(N+1):
        try:
            v0.append(u(i*w/N))
        except ZeroDivisionError:
            v0.append(-np.inf)
    k_list.append(np.argmax(v0))
    for t in range(1,T):
        v1 = []
        m = 0 
        maxval = 0.
        for i in range(N+1):
            v2 = []
            for j in range(i+1):
                try:
                    v2.append(u(j*w/N) + beta*v0[i-j])
                except ZeroDivisionError:
                    v2.append(-np.inf)
            v1.append(max(v2))
            if max(v2) >= maxval:
                m = np.argmax(v2)
                maxval = max(v2)
        k_list.append(m)
        v0 = v1
    c = []
    for k in k_list[::-1]:
        c.append(k*w/N)
        w -= k*w/N
    return c, np.max(v0)
```

```python
f = lambda x: np.log(x)
beta = 0.5
w = 100
T = 10
for n in range(100,500,100):
    print(cake_eating(w, beta, n, T, f))
    print()
```

## Problem 16.5 ii)

```python
def analytic(W, T, beta):
    u = lambda x: (x-1)/x
    c0 = W/sum(beta**(t/2) for t in range(T))
    c_t = lambda t: beta**(t/2)*c0
    c = [c0] + [c_t(t) for t in range(1,T)]
    V = sum(beta**i * u(ci) for i,ci in enumerate(c))
    return c, V
```

```python
u = lambda x: (x-1)/x
T = 15
N = 50
beta = 0.5
for W in range(20,50,10):
    print(analytic(W,T,beta))
    print(cake_eating(W, beta, N, T, u))
    print()
```

From these results, we can see that the analytic solutions and the value iteration solutions are fairly close. It is a lot closer as N gets larger, as shown below:

```python
u = lambda x: (x-1)/x
T = 15
N = 500
beta = 0.5
for W in range(20,50,10):
    print(analytic(W,T,beta))
    print(cake_eating(W, beta, N, T, u))
    print()
```
