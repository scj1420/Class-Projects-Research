```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Problem 9.17

```python
p_1 = lambda x: -27*np.sqrt(3)/32*((x+1)*(x-1/3)*(x-1)+(x+1)*(x+1/3)*(x-1))
x = np.linspace(-1,1,100)
plt.plot(x, np.sin(np.pi*x), label="sin(pi*x)")
plt.plot(x, p_1(x), label="9.17")
plt.legend()
plt.show()
```

## Problem 9.18

```python
p_2 = lambda x: -4/3*((x-1)*(x-1/2)*(x+1)+(x-1)*(x+1/2)*(x+1))
plt.plot(x, np.sin(np.pi*x), label="sin(pi*x)")
plt.plot(x, p_1(x), label="9.17")
plt.plot(x, p_2(x), label="9.18")
plt.legend()
plt.show()
```
