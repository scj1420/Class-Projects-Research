```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Problem 8.29

#(This is a demo plot)

def haar_son(j,k,x):
    if x >= k/(2**j) and x < (k+1)/(2**j):
        return 1
    else:
        return 0

def sample(f, n):
    k = np.linspace(0, 1, 2**n, endpoint=False)
    return f(k)
def func(a):
    n = int(np.log2(len(a)+1))
    return lambda x: sum([a_k*haar_son(n, k, x) for k,a_k in enumerate(a)])
```

```python
f = lambda t: np.sin(2*np.pi*t - 5)/(np.sqrt(abs(t - np.pi/20)))

for n in range(1,11):
    t = np.linspace(0, 1, 300, endpoint=False)
    f_n = func(sample(f,n))
    
    plt.plot(t, f(t))
    plt.plot(t,[f_n(t_k) for t_k in t])
    plt.show()
```
