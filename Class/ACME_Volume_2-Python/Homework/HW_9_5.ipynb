```python
# your code here, per problem
def cheb_interp(n):
    extrema = np.cos((np.pi*np.arange(2*n))/n)
    samples = func(extrema)
    coeffs = np.real(fft.fft(samples))[:n+1]
    coeffs = coeffs/n
    coeffs[0] = coeffs[0]/2
    coeffs[n] = coeffs[n]/2
    return coeffs
```

```python
def func(x):
    y = x.copy()
    y[y < 0] += 1
    y[y == 0] = 0
    return y
```

```python
x = np.linspace(-1,1,100)
np.polynomial.chebyshev.chebval(x, cheb_interp(128))
```

```python
for i in range(1,8):
    plt.plot(x, func(x))
    plt.plot(x, np.polynomial.chebyshev.chebval(x, cheb_interp(2**i)))
    plt.title("n = " + str(i))
    plt.show()
```
