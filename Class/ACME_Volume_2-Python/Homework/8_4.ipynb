```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sampling(v, n, T, f):
    x = np.linspace(0, T, n)
    f_dft = sp.fft(f(x))/n
    c_k = np.concatenate([f_dft[n-v:n], f_dft[:v+1]])
    return c_k

def gfunc(c_k, v, T):
    return lambda x: sum([c*np.exp(2*np.pi*1j*(k-v)*x/T) for k,c in enumerate(c_k)])
```

```python
f = lambda x: 1 - 3*np.sin(12*np.pi*x + 7) + 5*np.sin(2*np.pi*x - 1) + 5*np.sin(4*np.pi*x - 3)

T = 1
v = 6

for n in [3, 7, 11, 13, 100]:
    x = np.linspace(0, T, 200)
    v = n//2
    c = sampling(v,n,T,f)
    g = gfunc(c, v, T)
    plt.plot(x, f(x))
    plt.plot(x, np.real(g(x)))
    plt.show()
```
