```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python

def sampling(v, n, T, f):
    x = np.linspace(0, T, n)
    f_1 = sp.fft(f(x))/n
    c_k = np.hstack([f_1[n-v:n], f_1[0:v+1]])
    return c_k

def g_x(c_k, v, T):
    return lambda x: sum([c*np.exp(2*np.pi*1j*(k-v)*x/T) for k,c in enumerate(c_k)])
    
```

```python
f = lambda x: 1 - 3*np.sin(12*np.pi*x + 7) + 5*np.sin(2*np.pi*x - 1) + 5*np.sin(4*np.pi*x - 3)

v = 6
T = 1

for n in [3, 7, 11, 13]:
    v = n//2
    c_k = sampling(v, n, T, f)
    t = np.linspace(0, 1, 100)
    g = g_x(c_k, v, T)
    g = np.real(g(t))
    plt.plot(t, f(t))
    plt.plot(t, g)
    plt.title("n = " + str(n))
    plt.show()

```
