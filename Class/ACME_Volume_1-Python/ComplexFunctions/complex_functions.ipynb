## Problem 1
- Implement `plot_complex()`.
- Plot $f(z) = z$ on $\{x + yi \mid x,y \in [-1,1]\}$.
- Plot $f(z) = \sqrt{z^2 + 1}$ on $\{x + iy \mid x,y \in [-3,3]\}$.

```python
# Problem 1
def plot_complex(f, window=[-1, 1, -1, 1], res=400, title=""):
    """Visualize the complex-valued function f:C->C on the (res x res) domain
    {a + bi | a in [r_min, r_max], b in [i_min, i_max]} by plotting the
    argument of f(z) and the magnitude of f(z) in separate subplots.
    
    Parameters:
        f (func): A function that accepts and returns arrays with complex entries.
        window ([r_min, r_max, i_min, i_max]): The real and imaginary domain bounds.
        res (int): The plot resolution. The domain should be res x res.
        title (str): A label for the function to set as the figure title.
            Use r"$<latex code>$" for pretty printing. For example, for the function
            f = lambda z: z**2 - np.sqrt(z), use title=r"$f(z) = z^2 - \sqrt{z}$".
    """
    x = np.linspace(window[0], window[1], res) # Real domain
    y = np.linspace(window[2], window[3], res) # Imaginary domain
    X, Y = np.meshgrid(x, y) # Make grid matrices
    Z = X + 1j*Y # Combine the grids into a complex array
    
    # Plot the arg(f(z)) in a subplot
    plt.subplot(121)
    plt.pcolormesh(X, Y, np.angle(f(Z)), cmap="hsv", vmin=-np.pi, vmax=np.pi)
    plt.title("arg(f(z))")
    plt.gca().set_aspect('equal', adjustable='box') # Make sure the aspect ratio is equal
    plt.suptitle(title)
    
    #Plot the |f(z)| in a subplot
    plt.subplot(122)
    plt.pcolormesh(X, Y, np.abs(f(Z)), cmap="magma", norm=matplotlib.colors.LogNorm())
    plt.title("|f(z)|")
    plt.gca().set_aspect('equal', adjustable='box') # Make sure the aspect ratio is equal
    plt.suptitle(title)
    
    plt.show()
    
```

```python
# Plot the specified functions.
# Define the functions 
f1 = lambda z: z
f2 = lambda z: np.sqrt(z**2 + 1)
# Plot the functions
plot_complex(f1, title="f(z) = z")
plot_complex(f2, [-3, 3, -3, 3], title="f(z) = sqrt(z^2 + 1)")
```

## Problem 2

Plot the following functions on $\{x+iy \mid x \in [-1,1],\; y \in [-1,1]\}$.

- $f(z) = z^n$ for $n=2,3,4$.
- $f(z) = z^3 - iz^4 - 3z^6$.

What does a zero of $f(z)$ look like in these plots?

```python
# Problem 2
# Define and plot the z^n functions for each n
for n in [2,3,4]:
    fn = lambda z: z**n
    plot_complex(fn, title="f(z) = z^" + str(n))
# Define and plot the other function
f = lambda z: z**3 - 1j*z**4 - 3*z**6
plot_complex(f, title="f(z) = z^3 - iz^4 - 3z^6")
```

Answer: In the angle plot, the zeros appear as the focal point of the changing colors, so in an open small neighborhood around the zeros, different colors are contained within it even as the radius becomes very small. In the magnitude plot, the zeros appear as (blue) dots on the graph. The multiplicity can only be seen in the angle graph, and you can see it by how many times a certain color appears in a neighborhood around a zero. For example, in the $f(z) = z^3 - iz^4 - 3z^6$ graph, we can we that the point $z_0 = 0$ is a zero of multiplicity 3 because the color green appears 3 times around 0.

## Problem 3

Plot the following functions on domains that show all of the zeros and poles.

- $f(z) = z^{-n}$ for $n=1,2,3$.
- $f(z) = z^2 + iz^{-1} + z^{-3}$.

What does a pole of $f(z)$ look like in these plots?

```python
# Problem 3
# Define and plot the z^-n functions
for n in [2,3,4]:
    fn = lambda z: z**-n
    plot_complex(fn, title="f(z) = z^-" + str(n))
# Define and plot the other function
f = lambda z: z**2 + 1j*z**(-1) + z**(-3)
plot_complex(f, [-1.5, 1.5, -1.5, 1.5], title="f(z) = z^2 + iz^-1 + z^-3")
```

Answer: It seems that in the angle graph, the poles and the zeros appear identically as the focal points of the colors explained above. In order to distinguish the two, we must look at the magnitude graph. The poles appear as white dots, where as the zeros appear as black dots. The multiplicity of the poles can be determined the same way as the multiplicity of a zero. (number of times a certain color appears in a neighborhood)

## Problem 5

Plot the following functions on the domain $\{x+iy\mid x,y\in[-8,8]\}$.
Explain carefully what each graph reveals about the function and why the function behaves that way.

- $f(z) = e^z$.
- $f(z) = \tan(z)$.

(Hint: use the polar coordinate representation to mathematically examine the magnitude and angle of each function.)

```python
# Problem 5
# Define the functions
f1 = lambda z: np.exp(z)
f2 = lambda z: np.tan(z)
# Plot the functions
plot_complex(f1, [-10,10,-10,10], title="f(z) = e^z")
plot_complex(f2, [-10,10,-10,10], title="f(z) = tan(z)")
```

Answers:

$f(z) = e^z$

Let $z$ be a complex value, and let $z = x + iy$ where $x,y \in \mathbb{R}$, so $x = \Re(z)$ and $y = \Im(z)$. Also, let $f(z) = re^{i\theta}$ where $r,\theta \in \mathbb{R}$, so $r = |f(z)|$ and $\theta = arg(f(z))$. Then, we have $f(z) = e^z = e^{x+iy} = e^x e^{iy} = re^{i\theta}$. Therefore, $r = e^x$ and $\theta = y$. We can see this on the graph, that the angle of $f(z)$ solely depends on the imaginary part of $z$ and cycles through $2\pi$ as we would expect. The modulus of $f(z)$ solely depends on the real part of $z$, approaching $\infty$ as $x \rightarrow \infty$ and approaching $0$ as $x \rightarrow -\infty$

$f(z) = \tan(z)$

Since $f(z) = \tan(z) = \frac{\sin(z)}{\cos(z)}$, there are zeros each time $sin(z) = 0$, so whenever $z = n\pi, n \in \mathbb{Z}, f(z) = 0$, and there are poles each time $cos(z) = 0$, so whenever $z = n\pi + \frac{\pi}{2}, f(z) = \infty$. This can be clearly seen in the graph for $|f(z)|$.


## Problem 6

For each of the following functions, plot the function on the domain $\{x+iy\mid x,y\in[-1,1]\}$ and describe what this view of the plot seems to imply about the function.
Then plot the function on a domain that allows you to see the true nature of the roots and poles and describe how it is different from what the original plot implied.

- $f(z) = 100z^2 + z$
- $f(z) = \sin\left(\frac{1}{100z}\right)$.

(Hint: zoom way in.)

```python
# Problem 6
# Define the functions
f1 = lambda z: 100*z**2 + z
f2 = lambda z: np.sin(1/(100*z))
# Plot the functions over the domain [-1 1] x [-1 1]
plot_complex(f1, title="f(z) = 100z^2 + z")
plot_complex(f2, title="f(z) = sin(1/100z)")
# Plot the functions in a much more zoomed in domain. 
plot_complex(f1, [-2e-2, 1e-2, -1e-2, 1e-2], title="f(z) = 100z^2 + z, zoomed in")
plot_complex(f2, [-5e-3, 5e-3, -5e-3, 5e-3], title="f(z) = sin(1/100z), zoomed in")
```

Answers: When plotted on the window $\{x + iy\ |\ x,y \in [-1, 1]\}$, the first function just seem to have a zero of multiplicity 2 at the origin and the second function just seem to have a pole of multiplicity 1 at the origin. However, when zoomed in, we can clearly see that the first function has 2 zeros of multiplicity 1, one at $z = -\frac{1}{100}$ and the other at the origin, and the second function has an essential pole at the origin.
